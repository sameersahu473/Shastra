```python
def transform(str):
  if len(str)== 0 :
     return 0
  
  else:
    if len(str) ==3:
      return int(str[0:1])
    else:
     return int(str[0:2])

     
```

```python
def convert_attributes(number_str):
  if type(number_str) == str:
    if '+' in number_str:
      return float(number_str.split('+')[0])
    elif '-' in number_str:
      return float(number_str.split('-')[0])
    else:
      return float(number_str)
```

```python
def splitter(a):
  ans = []
  i = 0
  while i < len(a):
    temp = 0
    while a[i]>='0' and a[i]<='9':
      temp = int(temp*10) + int(a[i])
      i+=1
    if temp>0:
      ans.append(temp)
    i+=1
  return ans
```

```python
x = splitter(df_id.iloc[0][0])
print(x)
```

```python
z = splitter(df_id.iloc[0][0])
val = df1[df1['sofifa_id']==z[0]]['predictions']
print(val)
```

```python
ans =[]
for row in range(len(df_id)):
  index=-1
  val_max = 0
  x = splitter(df_id.iloc[row][0])

  for i in range(len(x)):
    val = df1[df1['sofifa_id']==x[i]]['predictions'].iloc[0]
    if val> val_max:
      val_max = val
      index = i
  ans.append(x[index])
print(ans)
```

```python
y_ans = np.array(ans)
```
